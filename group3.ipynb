```python
from sklearn.model_selection import train_test_split

train,test= train_test_split(data_raw
                             , test_size=0.20,random_state=7)
```

```python
import random
data_F = pd.DataFrame(train[train['FLAG'] == 1])
F1 = sum(train['FLAG'] == 1)
F0 = sum(train['FLAG'] == 0)
train_re = pd.DataFrame(train)
train_re = train_re.append(data_F.sample(n = F0-F1, replace=True, random_state=1))
```

```python
from sklearn import svm, tree, linear_model, neighbors, naive_bayes, ensemble, discriminant_analysis, gaussian_process
from xgboost import XGBClassifier

#Common Model Helpers
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn import feature_selection
from sklearn import model_selection
from sklearn import metrics

#Visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
from pandas.plotting import scatter_matrix

MLA = [
    #Ensemble Methods
    ensemble.AdaBoostClassifier(),
    ensemble.BaggingClassifier(),
    ensemble.ExtraTreesClassifier(),
    ensemble.GradientBoostingClassifier(),
    ensemble.RandomForestClassifier(),

    #Gaussian Processes
    gaussian_process.GaussianProcessClassifier(),
    
    #GLM
    linear_model.LogisticRegressionCV(),
    linear_model.PassiveAggressiveClassifier(),
    linear_model.RidgeClassifierCV(),
    linear_model.SGDClassifier(),
    linear_model.Perceptron(),
    
    #Navies Bayes
    naive_bayes.BernoulliNB(),
    naive_bayes.GaussianNB(),
    
    #Nearest Neighbor
    neighbors.KNeighborsClassifier(),
    
    #SVM
    svm.SVC(probability=True),
    svm.NuSVC(probability=True),
    svm.LinearSVC(),
    
    #Trees    
    tree.DecisionTreeClassifier(),
    
    #Discriminant Analysis
    discriminant_analysis.LinearDiscriminantAnalysis(),
    discriminant_analysis.QuadraticDiscriminantAnalysis(),

    
    #xgboost: http://xgboost.readthedocs.io/en/latest/model.html
    XGBClassifier()    
    ]
```

```python

cv_split = model_selection.ShuffleSplit(n_splits = 10, test_size = .3, train_size = .6, random_state = 0 ) # run model 10x with 60/30 split intentionally leaving out 10%

#create table to compare MLA metrics
MLA_columns = ['MLA Name', 'MLA Parameters','MLA Train Accuracy Mean', 'MLA Test Accuracy Mean','MLA Test Power', 'MLA Test Accuracy 3*STD' ,'MLA Time']

MLA_compare = pd.DataFrame(columns = MLA_columns)
#create table to compare MLA predictions
MLA_predict = train_re['FLAG']
#index through MLA and save performance to table
row_index = 0
for alg in MLA:

    #set name and parameters
    MLA_name = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Name'] = MLA_name
    MLA_compare.loc[row_index, 'MLA Parameters'] = str(alg.get_params())
    cv_results = model_selection.cross_validate(alg, train_re[[i for i in list(train_re.columns) if i != 'FLAG']], train_re['FLAG'], cv  = cv_split, return_train_score=True)
    
    #print(cv_results)
    MLA_compare.loc[row_index, 'MLA Time'] = cv_results['fit_time'].mean()
    MLA_compare.loc[row_index, 'MLA Train Accuracy Mean'] = cv_results['train_score'].mean()
    MLA_compare.loc[row_index, 'MLA Test Accuracy Mean'] = cv_results['test_score'].mean()   
    MLA_compare.loc[row_index, 'MLA Test Accuracy 3*STD'] = cv_results['test_score'].std()*3   #let's know the worst that can happen!
    

    #save MLA predictions - see section 6 for usage4
    alg.fit(train_re[[i for i in list(train_re.columns) if i != 'FLAG']],train_re['FLAG'])
    MLA_predict[MLA_name] = alg.predict(train_re[[i for i in list(train_re.columns) if i != 'FLAG']])
    row_index+=1
    
    #print and sort table: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sort_values.html
    MLA_compare.sort_values(by = ['MLA Test Accuracy Mean'], ascending = False, inplace = True)
    MLA_compare
```

```python
MLA_compare.sort_values(by = ['MLA Test Accuracy Mean'], ascending = False, inplace = True)
sns.barplot(x='MLA Test Accuracy Mean', y = 'MLA Name', data = MLA_compare,color = 'blue')

plt.title('Machine Learning Algorithm Accuracy Score \n')
plt.xlabel('Accuracy Score (%)')
plt.ylabel('Algorithm')
```

```python
X_train = train_re[[i for i in list(train_re.columns) if i != "FLAG"]]
Y_train = train_re['FLAG']
X_test = test[[i for i in list(test.columns) if i != "FLAG"]]
Y_test = test['FLAG']
```

```python
MLA_name = ['RandomForest','ExtraTrees','XGB','Bagging','GaussianProcess','DecisionTree','GradientBoosting','AdaBoost']
acc = [0]*8
pow = [0]*8
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
nlist = [10,30,50,80,100,150,200,300,500]
```

```python
#Random Forest
RF_par = {'n_estimators': nlist, 'min_samples_leaf': [1, 2, 4], 'min_samples_split': [2, 5, 10]}

RF_grid = GridSearchCV(ensemble.RandomForestClassifier(), RF_par, cv=5, n_jobs=-1, verbose=True)

RF_grid.fit(X_train, Y_train)

print (f"Best parameters are {RF_grid.best_params_}") 

pre = RF_grid.predict(X_test)

acc[0] = sum(pre == Y_test)/len(Y_test)
pow[0] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
#Extra Trees
ET_par = {'n_estimators': nlist, 'min_samples_leaf': [1, 2, 4], 'min_samples_split': [2, 5, 10], 'max_depth':[3,5,10,30]}

ET_grid = GridSearchCV(ensemble.ExtraTreesClassifier(), ET_par, cv=5, n_jobs=-1, verbose=True)

ET_grid.fit(X_train, Y_train)

print (f"Best parameters are {ET_grid.best_params_}") 

pre = ET_grid.predict(X_test)

acc[1] = sum(pre == Y_test)/len(Y_test)
pow[1] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
#XGboost
XGB_par = {'n_estimators': nlist, 'max_depth':[3,5,10,30],'learning_rate':[0.01,0.1,1]}

XGB_grid = GridSearchCV(XGBClassifier(), XGB_par, cv=5, n_jobs=-1, verbose=True)

XGB_grid.fit(X_train, Y_train)

print (f"Best parameters are {XGB_grid.best_params_}") 

pre = XGB_grid.predict(X_test)

acc[2] = sum(pre == Y_test)/len(Y_test)
pow[2] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
#Bagging
B_par = {'n_estimators': nlist, 'max_samples': [0.1, 0.5, 0.75, 1.0], 'max_features': [2, 5, 10]}

B_grid = GridSearchCV(ensemble.BaggingClassifier(), B_par, cv=5, n_jobs=-1, verbose=True)

B_grid.fit(X_train, Y_train)

print (f"Best parameters are {B_grid.best_params_}") 

pre = B_grid.predict(X_test)

acc[3] = sum(pre == Y_test)/len(Y_test)
pow[3] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
#Gaussian Process
GP =  gaussian_process.GaussianProcessClassifier()

GP.fit(X_train, Y_train)

pre = GP.predict(X_test)

acc[4] = sum(pre == Y_test)/len(Y_test)
pow[4] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
#Decision Tree
DT_par = {'max_features': [2, 5, 10],'max_depth':[5,10,30,50]}

DT_grid = GridSearchCV(tree.DecisionTreeClassifier(), DT_par, cv=5, n_jobs=-1, verbose=True)

DT_grid.fit(X_train, Y_train)

print (f"Best parameters are {DT_grid.best_params_}") 

pre = DT_grid.predict(X_test)

acc[5] = sum(pre == Y_test)/len(Y_test)
pow[5] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
#Gradient Boost
GB_par = {'n_estimators': nlist, 'min_samples_leaf': [1, 2, 4], 'min_samples_split': [2, 5, 10],'max_depth':[3,5,10,30]}

GB_grid = GridSearchCV(ensemble.GradientBoostingClassifier(), GB_par, cv=5, n_jobs=-1, verbose=True)

GB_grid.fit(X_train, Y_train)

print (f"Best parameters are {GB_grid.best_params_}") 

pre = GB_grid.predict(X_test)

acc[6] = sum(pre == Y_test)/len(Y_test)
pow[6] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
#AdaBoost
AB_par = {'n_estimators': nlist, 'learning_rate':[0.01,0.1,1,3,5]}

AB_grid = GridSearchCV(ensemble.AdaBoostClassifier(), AB_par, cv=5, n_jobs=-1, verbose=True)

AB_grid.fit(X_train, Y_train)

print (f"Best parameters are {AB_grid.best_params_}") 

pre = AB_grid.predict(X_test)

acc[7] = sum(pre == Y_test)/len(Y_test)
pow[7] = sum(np.array(pre == Y_test) & np.array(Y_test == 1))/sum(Y_test == 1)
```

```python
Performance = pd.DataFrame({'Name':MLA_name,'Accuracy':acc,'Power':pow})
Performance.sort_values(by = ['Power'], ascending = False, inplace = True)
Performance
```
